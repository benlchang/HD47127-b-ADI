```python
import pysynphot as S
import numpy as np
```

```python
w, t = np.loadtxt('2MASS_2MASS.H.dat', unpack=True)
```

```python
bp_H = S.ArrayBandpass(w, t, name='2mass')

w_CH4, t_CH4 = np.loadtxt('Keck_NIRC2.CH4s.csv', delimiter=',', unpack=True)
bp_ch4 = S.ArrayBandpass(w_CH4, t_CH4, name='ch4')
```

```python
w_model, t_model = np.loadtxt('model_spectrum.txt', unpack=True)

_, idx = np.unique(w_model, return_index=True)

sp = S.ArraySpectrum(w_model[idx], t_model[idx], fluxunits='flam', name='model')
```

```python
sp_norm = sp.renorm(5.277, 'vegamag', bp_H)
```

```python
obs = S.Observation(sp_norm, bp_ch4)
print(obs.effstim('vegamag'))
```
